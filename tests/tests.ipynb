```python
import numpy as np
import scipy.stats as scs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv('../Data/diamonds.csv')
```

```python
diamonds.head()
```

```python
sns.histplot(scs.norm.pdf(x=diamonds['carat'], loc=diamonds['carat'].mean(), scale=diamonds['carat'].std()))
```

```python
sns.histplot(scs.norm.cdf(x=diamonds['carat'], loc=diamonds['carat'].mean(), scale=diamonds['carat'].std()))
```
